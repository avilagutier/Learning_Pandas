```python
# Index: cambiar de nombre la columna Country al inicio del dataframe
df = pd.read_csv("../Data_Science_Alex/world_population.csv", index_col = 'Country')
df
```

```python
# Como invertir el datraframe
df.reset_index(inplace=True)
df
```

```python
df.set_index('Country', inplace = True) #Inplace = True, permite guardar el comando de nombre de la columna Country 
df
```

```python
# Forma comun de encontrar data en el dataframe

#loc 

df.loc['Albania'] # permite la localizacion de la columna del pais seleccionado, para ver todos sus datos 
```

```python
df.iloc[1] #iloc, permite seleccionar el numero de fila 
```

```python
df.iloc[4]
```

```python
df.loc['Andorra']
```

```python
# Multi-indexing
# Se tiene dos indexes obtenidos, Continent y Country
df.set_index(['Continent', 'Country'], inplace = True)
df
```

```python
# Mejorar estos valores de la tabla anterior, ordenando (sort) la tabla
df.sort_index()
```

```python
df.sort_index(ascending =False) # se ordena de forma alfabetica o no 
```

```python
df.sort_index(ascending =[False, True])
```

```python
df.loc['Angola'] # no funciona en este caso porque esta buscando con la primer columna 
```

```python
df.loc['Africa'] #Pone solo los paises de un solo continente, en este caso Africa 
```

```python
df.loc['Africa', 'Angola'] # selecciona primero todos los paises del continene de Africa y luego solo se selecciona el pais - Angloa 
```
